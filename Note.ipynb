# Chiến lược Customer 360 với dữ liệu Olist (Brazil)

Việc xây dựng bài toán **Customer 360** trên bộ dữ liệu Olist là một thử thách thú vị nhưng cũng đầy “chông gai” vì **tỷ lệ khách hàng quay lại rất thấp (~3% mua >1 lần)**.
Do đó, thay vì tập trung vào **Vòng đời khách hàng dài hạn (Long-term CLV)** như các mô hình thương mại điện tử thông thường, chiến lược Customer 360 tại đây cần tập trung vào:

* **Trải nghiệm tại thời điểm mua hàng**
* **Hành vi chi tiêu**
* **Sự hài lòng của khách hàng**

Dưới đây là bộ câu hỏi nghiên cứu được phân chia theo 4 cấp độ phân tích:

---

## 1. Descriptive Analysis (Phân tích Mô tả)

**Mục tiêu:** Vẽ chân dung khách hàng là ai và họ đã làm gì? (The "Who" and "What")
Với tỷ lệ khách hàng mua lại thấp, tập trung vào **snapshot tại thời điểm giao dịch**.

* **Tổng quan khách hàng**:

  * Phân bố theo địa lý (Bang, Thành phố)
  * Khu vực tập trung khách hàng giá trị cao (High value)

* **Hành vi thanh toán**:

  * Tỷ lệ sử dụng các phương thức thanh toán (Credit Card, Boleto, Voucher, Debit)
  * Xu hướng trả góp (installments) hay trả thẳng
  * Số kỳ trả góp trung bình

* **Giá trị đơn hàng**:

  * Phân phối giá trị đơn hàng (Basket Size)
  * Phí vận chuyển (Freight Value) và tỷ lệ phí trên tổng giá trị đơn hàng

* **Sản phẩm ưa thích**:

  * Danh mục sản phẩm thu hút khách hàng mới lớn nhất

* **Chất lượng dịch vụ**:

  * Điểm đánh giá trung bình (Review Score)
  * Tỷ lệ đánh giá 5 sao so với 1 sao

---

## 2. Exploratory Analysis (Phân tích Khám phá)

**Mục tiêu:** Tìm ra mối liên hệ giữa các biến số và lý giải hình dạng dữ liệu.

* **Địa lý & Chi phí**:

  * Vị trí khách hàng ảnh hưởng thế nào đến phí vận chuyển (Freight Value) và thời gian giao hàng (Delivery Time)?
  * Khách ở vùng xa (ví dụ: phía Bắc Brazil) có chịu phí cao hơn và chờ lâu hơn?

* **Thanh toán & Chi tiêu**:

  * Mối tương quan giữa phương thức thanh toán và giá trị đơn hàng?
  * Ví dụ: Người dùng thẻ tín dụng có xu hướng mua đơn giá trị cao hơn người dùng Boleto không?

* **Thời gian & Sự hài lòng**:

  * Thời gian giao hàng thực tế so với dự kiến ảnh hưởng thế nào đến Review Score?
  * Gap thời gian càng lớn thì điểm càng thấp?

* **Đặc tính sản phẩm**:

  * Kích thước/khối lượng sản phẩm ảnh hưởng thế nào đến hành vi mua hàng và phí vận chuyển?

* **Ngày/Giờ mua sắm**:

  * Khung giờ hoặc ngày nào trong tuần khách hàng đặt hàng nhiều nhất?

---

## 3. Diagnostic Analysis (Phân tích Chẩn đoán)

**Mục tiêu:** Trả lời câu hỏi “Tại sao?” và phân tích các pain points.

* **Nguyên nhân đánh giá thấp (Root Cause of Bad Reviews)**:

  * Phân tích text (NLP) trong comment
  * Do giao hàng chậm? Hàng hỏng? Sai mô tả?

* **Vấn đề giao vận**:

  * Vì sao một số khu vực có tỷ lệ giao hàng trễ cao bất thường?

* **Hủy đơn & Thất bại**:

  * Yếu tố nào thường xuất hiện với các đơn hàng bị hủy hoặc không giao thành công?

* **Phân tích sự rời bỏ (Implicit Churn Analysis)**:

  * Nhóm nhỏ khách quay lại (~3%) có gì khác biệt với nhóm còn lại 97%?
  * Thời gian trung bình giữa 2 lần mua, tốc độ giao hàng lần 1, v.v.

---

## 4. Predictive Analysis (Phân tích Dự báo)

**Mục tiêu:** Sử dụng **Clustering** và **Classification** để dự đoán hành vi và phân khúc khách hàng.
Do thiếu lịch sử giao dịch, tập trung vào **Giá trị và Sự hài lòng** thay vì dự đoán Churn.

### A. Clustering (Phân cụm khách hàng)

**Kỹ thuật:** K-Means, DBSCAN hoặc Hierarchical Clustering
**Câu hỏi:** Chia khách hàng thành các nhóm Persona dựa trên:

* **Recency:** Thời gian mua gần nhất
* **Monetary:** Giá trị đơn hàng
* **Freight Ratio:** Tỷ lệ phí ship/giá sản phẩm
* **Product Diversity:** Số lượng category khác nhau trong giỏ hàng

**Ví dụ Output:**

| Nhóm   | Mô tả                                                   |
| ------ | ------------------------------------------------------- |
| Nhóm 1 | "Đại gia sỉ" – Mua nhiều, giá trị cao, ở thành phố lớn  |
| Nhóm 2 | "Khách lẻ vùng sâu" – Mua ít, phí ship cao, chờ lâu     |
| Nhóm 3 | "Tín đồ trả góp" – Giá trị đơn cao, chia nhỏ thanh toán |

### B. Classification (Phân loại & Dự báo)

**Kỹ thuật:** Logistic Regression, Random Forest, XGBoost

* **Dự báo sự hài lòng (Satisfaction Prediction):**

  * Dựa trên thông tin đơn hàng (Sản phẩm, Phí ship, Thời gian giao hàng dự kiến, Khoảng cách địa lý)
  * Mục đích: Can thiệp sớm nếu mô hình dự báo khách không hài lòng

* **Dự báo khả năng quay lại (Repurchase Probability):**

  * Dựa trên đặc điểm đơn hàng đầu tiên, xác định khách hàng “tiềm năng quay lại”
  * Mục đích: Tối ưu ngân sách marketing, retarget nhóm có xác suất cao thay vì spam toàn bộ

---

## Tóm tắt chiến lược dữ liệu

* **Xử lý “One-time buyer”**: Không dùng mô hình CLV truyền thống, mỗi khách hàng là một observation về trải nghiệm mua sắm.
* **Tận dụng dữ liệu địa lý (Geolocation):** Logistic là yếu tố quan trọng tại Brazil.
* **Tận dụng NLP:** Dữ liệu review text quý giá để làm Diagnostic Analysis.
